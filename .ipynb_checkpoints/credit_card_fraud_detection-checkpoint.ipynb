```octave
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```octave
df = pd.read_csv(r"C:\\Users\\pavit\\Downloads\\credit_card_fraud_10k.csv")
df.head()
```

```octave
df.isnull().sum()
```

```octave
sns.countplot(x = 'is_fraud', data=df)
plt.show()
```

```octave
sns.lineplot(x='transaction_hour', y='is_fraud', data=df)
plt.show()
```

```octave
fraud_rate = (
    df.groupby('transaction_hour')['is_fraud']
      .mean()
      .reset_index()
)

plt.figure()
sns.barplot(
    data=fraud_rate,
    x='transaction_hour',
    y='is_fraud'
)
plt.ylabel("Fraud Rate")
plt.xlabel("Transaction Hour")
plt.show()
```

```octave
sns.lineplot(x = 'device_trust_score', y='is_fraud', data=df)
plt.show()
```

```octave
cond1 = ((df['device_trust_score'] < 50) & (df['is_fraud'] == 1)).sum()
cond2 = (df['is_fraud'] == 1).sum()
print(cond1 / cond2)
```

```octave
plt.figure()
sns.boxplot(x='is_fraud', y='amount', data=df)
plt.show()
```

```octave
print(f"Amount greater than 500 bucks and is fraud : {((df['amount'] >= 500) & (df['is_fraud'] == 1)).sum()}")
print(f"Average transaction amount : {df['amount'].mean()}")
```

Hence from this we can conclude that the device_trust_score column has much higher linear correlation with is_fraud

```octave
print(f"{((df['foreign_transaction'] == 1) & (df['location_mismatch'] == 1) & (df['is_fraud'] == 1)).sum()}")
```

Hence we can also say that if the transaction is foreign and location mismatch then is_fraud. This statement is true upto some extent but cannot be said that it is always true.

```octave
plt.figure()
sns.histplot(
    data=df,
    x='velocity_last_24h',
    hue='is_fraud',
    bins=30,
    element='step',
    stat='density',
    common_norm=False
)
plt.show()
```

Velocity_last_24h which represents number of transactions attempted in the last 24hours. One of the high ROI feature for is_fraud

```octave
num_df = df.select_dtypes(include=['int64', 'float64'])
corr = num_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    center=0,
    linewidths=0.5
)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```octave
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, average_precision_score
X = df.drop(columns=['is_fraud', 'merchant_category'])
y = df['is_fraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state = 40)
```

```octave
print(y.value_counts(normalize=True))
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
```

```octave
rf = RandomForestClassifier(
    n_estimators = 200,
    max_depth = None,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight='balanced',
    random_state=40,
    n_jobs= -1
)
rf.fit(X_train, y_train)
```

```octave
y_prob = rf.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_prob)
pr_auc  = average_precision_score(y_test, y_prob)

print("ROC-AUC:", roc_auc)
print("PR-AUC :", pr_auc)
```
